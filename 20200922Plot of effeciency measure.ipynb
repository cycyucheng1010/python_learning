```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
#程式碼參考 https://gist.github.com/niftycode/917bb33be4e91a10b1fffd020f987f94
#niftycode
'''
big-o-notation.py:
Version: 0.2
Python 3.6
Date created: 22/03/2017
'''

# Big-O   	Name
# 1   		Constant
# log(n)  	Logarithmic
# n   		Linear
# nlog(n) 	Log Linear
# n^2 		Quadratic
# n^3 		Cubic
# 2^n 		Exponential

import numpy as np
import matplotlib.pyplot as plt


# Stylesheets defined in Matplotlib
plt.style.use('seaborn-darkgrid')

# Set up runtime comparisons
n = np.linspace(1, 10, 1000)#1~10之間1000點數據
labels = ['Constant', 'Logarithmic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic', 'Exponential']
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]

# Plot setup
plt.figure(figsize=(12, 10))#寬跟長
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('Input Size')
plt.savefig('big-o-notation.png')
```

```
#程式碼參考 https://gist.github.com/niftycode/917bb33be4e91a10b1fffd020f987f94
# author: niftycode
#使用工具:google colab
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')        #使用之布景樣式
n = np.linspace(1, 10, 1000)          #生出範圍區間內取之等差數值
labels = [ 'Logarithmic(log(n))', 'Linear(n)', 'Log_Linear(n *log(n))', 'Quadratic(n^2)', 'Cubic(n^3)', 'Exponential(2^n)']
big_o = [ np.log(n), n, n * np.log(n), n**2, n**3, 2**n]      #利用np將n帶入公式求點

# Plot setup
plt.figure(figsize=(12, 10))          #圖形的寬跟長
plt.ylim(0, 50)                 #設定y的上限與下限

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])  #將點連線畫出圖形
plt.legend()      #圖例

plt.ylabel('O(n)')  #ybar
plt.xlabel('n')    #xbar
plt.savefig('plot of Effeciency Measures.png')  #存圖
```
